```python
# mendrop missing valuesnya
# men-drop baris di kolom Embark yang memuat missing values

data = data.dropna(axis=0, subset=['Embarked'])
```

```python
rerata_usia = data['Age'].mean()
```

```python
data['Age_fillna_mean'] = data['Age'].fillna(rerata_usia)
```

```python
data['Cabin_fillna'] =  data['Cabin'].fillna('Tidak Punya Kabin')
```

```python
# Formatting
data['Age_fillna_mean'] = data['Age_fillna_mean'].round(0)
```

```python
# konversi float/desimal ke bilangan bulat
data['Age_fillna_mean'] = data['Age_fillna_mean'].astype(int)
```
