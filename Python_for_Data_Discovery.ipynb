```python
# mendrop missing valuesnya
# men-drop baris di kolom Embark yang memuat missing values

data = data.dropna(axis=0, subset=['Embarked'])
```

```python
data['Age_fillna_mean'] = data['Age'].fillna(rerata_usia)
```

```python
data['Cabin_fillna'] =  data['Cabin'].fillna('Tidak Punya Kabin')
```

```python
# Formatting
data['Age_fillna_mean'] = data['Age_fillna_mean'].round(0)
```

```python
data.info()
```

```python
# konversi float/desimal ke bilangan bulat
data['Age_fillna_mean'] = data['Age_fillna_mean'].astype(int)
```

```python
data.info()
```

```python
data.head()
```

```python
# mengganti case dari teks

data['Name_upper'] = data['Name'].str.upper()
data['Name_lower'] = data['Name'].str.lower()
data['Name_title'] = data['Name'].str.title()
```

```python
data.head()
```

```python
# binning/pengelompokan

data['Age_category'] = pd.cut(
    data['Age'],
    [0,5,11,16,25,35,45,55,65,100],
    labels=['Balita','Kanak-kanak','Remaja Awal','Remaja Akhir', 'Dewasa Awal', 'Dewasa Akhir','Lansia Awal','Lansia Akhir','Manula']
    )
```

```python
data.head()
```

```python
# membuat dictionary yang memetakan tiap kelompok

age_category = {
    'Anak-anak':['Balita','Kanak-kanak'],
    'Remaja':['Remaja Awal','Remaja Akhir'],
    'Dewasa':['Dewasa Awal','Dewasa Akhir'],
    'Lansia':['Lansia Awal','Lansia Akhir','Manula']
}
```

```python
# konsep dictionary
# list comprehension

age_map = {k: v for v, lst in age_category.items() for k in lst}
data['Age_new_category'] = data['Age_category'].map(age_map)
```

```python
age_map
```

```python
data.head()
```

```python
# mereplace nilai

data['Emabarked_full'] = data['Embarked'].replace(
    {
        'S':'Southampton',
        'C':'Cherbourg',
        'Q':'Queenstown'
    }
)
```
